### 1.2 「次のlist_1をデータフレームに変換し、列名をそれぞれ['名前', '職業', '年齢', '身長', '体重']としてください。また、そのデータフレームをdf_1という名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「Series型, DataFrame型の作成方法」節
</details>

```python
# こちらのセルを実行してください。
list_1 = [['たかし', '学生', 17, 170, 60],
          ['かな', '学生', 15, 150, 43],
          ['はると', '社会人', 25, 172, 65],
          ['もも', '社長', 30, 158, 48],
          ['まい', '社会人', 23, 163, 51],
          ['おさむ', '専門学生', 19, 175, 68],
          ['あきら', '大学生', 20, 169, 63]]

```

```python
df_1 = pd.DataFrame(list_1, columns = ['名前', '職業', '年齢', '身長', '体重'])
df_1
```

### 1.3 「1.2と同様に、次のlist_2をデータフレームに変換し、列名をそれぞれ['名前', '職業', '年齢', '身長', '体重']としてください。また、そのデータフレームをdf_2という名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「Series型, DataFrame型の作成方法」節
</details>

```python
# こちらのセルを実行してください。
list_2 = [['そら', np.nan, 20, 173, 61],
          ['あやか', '学生', 19, 165, np.nan]]
```

```python
df_2 = pd.DataFrame(list_2, columns =['名前', '職業', '年齢', '身長', '体重'])
df_2
```

### 1.4 「df_1とdf_2を縦に結合したデータフレームをdf_concatenatedという名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「テーブル結合方法」節
</details>

```python
df_concatenated = pd.concat([df_1, df_2])
df_concatenated
```

### 1.5 「df_concatenatedを年齢・身長の順にどちらも降順で並び替えてください。これは年齢を降順に並べ替え、年齢が同じ時は身長で降順に並べると言うことです。また、そのデータフレームを同名の変数に再代入して、表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「加工処理」節
</details>

```python
df_concatenated = df_concatenated.sort_values(['年齢', '身長'], ascending=[False, False])
print(df_concatenated)
```

### 1.6 「df_concatenatedのインデックスを0から順に振り直して、同名の変数に再代入して表示してください。その際に、元のインデックスを保持する必要はありません。」

</details>

```python
df_concatenated = df_concatenated.reset_index(drop=True)
display(df_concatenated)
```

### 1.12 「次のlist_3をデータフレームに変換し、列名をそれぞれ['名前', '性別']としてください。また、そのデータフレームをdf_3という名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「Series型, DataFrame型の作成方法」節
</details>

```python
# こちらのセルを実行してください。
list_3 = [['たかし', '男性'], ['かな', '女性'], ['はると', '男性'], ['もも', '女性'], ['まい', '女性'], ['そら', '男性'], ['あきら', '男性']]
```

```python
list_3 = pd.DataFrame(list_3, columns = ['名前', '性別'])
df_3 = list_3
df_3
```

### 1.13 「df_concatenatedとdf_3を「名前」列で横に結合してください。結合の際には、どちらものデータフレームの全ての行を全て使って、結合してください。その横に結合したデータフレームをdf_mergedという名前の変数に代入して、表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「テーブル結合方法」節
</details>

```python
df_merged = pd.merge(df_concatenated, df_3, how = 'outer')
df_merged
```

### 1.14 「df_mergedの中で、欠損値が含まれている行を全て削除し、df_merge_droppedという名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「加工処理」節
</details>

```python
df_merge_dropped = df_merged.dropna()
df_merge_dropped
```

### 1.15 「df_merge_droppedに、「性別」列が男性なら0・女性なら1のように対応させた「性別_int」列を新たに追加してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「テキストデータの加工」節
</details>

```python
df_merge_dropped['性別_int'] = df_merge_dropped['性別'].map({'男性': 0, '女性': 1})
df_merge_dropped
```

### 1.16 「df_merge_droppedの「職業」列に「学生」という文字が含まれている行を全て抽出して、df_studentsという名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「テキストデータの加工」節
</details>

```python
df_students = df_merge_dropped[df_merge_dropped['職業'].str.contains('学生')]
df_students
```

### 1.17 「df_studentsの「身長」列の最大値を表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「要約統計量算出方法」節
