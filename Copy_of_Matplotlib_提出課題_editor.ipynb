```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

### 1.3 「次のコードを実行してください。」

```python
# こちらのセルを実行してください。
data = pd.DataFrame()
np.random.seed(100)
data['身長'] = np.random.normal(170, 7, 100)
data['クラス'] = ['a', 'b', 'c', 'd'] * 25
data['成績'] = [2, 5, 4, 1, 4, 3, 4, 2, 4, 5] * 10
data['体重'] = np.random.normal(60, 3, 100)

data
```

### 1.4 「dataの『体重』列の分布を、箱ひげ図によって可視化してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」節
</details>

```python
plt.figure(figsize=(10, 5))
plt.boxplot(data['体重'])
plt.show()
```

### 1.5 「クラス別の『体重』列の平均値を、棒グラフによって可視化してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」節
</details>

```python
plt.figure(figsize=(10, 5))
mean_weights = data.groupby('クラス')['体重'].mean()
plt.bar(mean_weights.index, mean_weights)
plt.show()
```

### 1.8 「1.7の散布図と同様のものを描画してください。その際に、タイトルを『身長と体重の関係』、x軸のラベルを『身長』、y軸のラベルを『体重』のように指定してください。」

<details>

### 1.9 「x軸にインデックス、y軸に身長を描画した折れ線グラフを描画してください。その際に、グラフはクラスごとのグラフにして、それぞれに組名のラベルをつけて、凡例を付与してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」・「各種グラフのオプション設定方法」節

```python
plt.figure(figsize=(10, 5))
for class_name, group in data.groupby('クラス'):
    plt.plot(range(25), group['身長'], label=class_name)
plt.legend()
plt.show()
```

### 1.10 「dataの『身長』列の分布を確認するために、クラスA・クラスBの『身長』列についてそれぞれのヒストグラムを描画してください。その際に、グラフの透明度を半透明にしてください。(透明度は好きな値に設定してください。)」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」・「各種グラフのオプション設定方法」節
</details>

```python
plt.figure(figsize=(10, 5))
for class_name, group in data.groupby('クラス'):
    plt.hist(group['身長'], alpha=0.4, label=class_name)

plt.title('身長')
plt.xlabel('身長')
plt.ylabel('人数')
plt.legend()
plt.show()
```
